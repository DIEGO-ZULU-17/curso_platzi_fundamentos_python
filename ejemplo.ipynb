```python
def clasificar_triangulo():
    """
    Esta función solicita al usuario tres longitudes y determina:
    1. Si pueden formar un triángulo válido
    2. Qué tipo de triángulo es (equilátero, isósceles o escaleno)
    3. El perímetro del triángulo
    4. Información adicional según el tipo
    """
    
    # === PASO 1: Solicitar datos al usuario ===
    # Usamos input() para que el usuario escriba valores
    # float() convierte el texto en número decimal
    try:
        lado1 = float(input("Ingrese la longitud del lado 1: "))
        lado2 = float(input("Ingrese la longitud del lado 2: "))
        lado3 = float(input("Ingrese la longitud del lado 3: "))
    except ValueError:
        print("Error: Debe ingresar números válidos.")
        return  # Termina la función si hay error
    
    # === PASO 2: Verificar que las longitudes sean positivas ===
    # Un triángulo no puede tener lados negativos o cero
    if lado1 <= 0 or lado2 <= 0 or lado3 <= 0:
        print("Error: Todas las longitudes deben ser números positivos.")
        return # Sale de la función si alguno de los lados es cero o negativo, ya que un triángulo no puede tener longitudes no positivas, y se informa al usuario del error para que proporcione valores válidos.
    
    # === PASO 3: Verificar la desigualdad triangular ===
    # Para que tres longitudes formen un triángulo, se debe cumplir:
    # La suma de cualquier dos lados debe ser mayor que el tercer lado
    if (lado1 + lado2 <= lado3) or (lado1 + lado3 <= lado2) or (lado2 + lado3 <= lado1):
        print(f"\nCon los valores {lado1}, {lado2} y {lado3} NO se puede formar un triángulo.")
        print("Recuerda: La suma de dos lados cualesquiera debe ser mayor que el tercer lado.")
        return
    
    # === PASO 4: Calcular el perímetro ===
    # El perímetro es la suma de los tres lados
    perimetro = lado1 + lado2 + lado3
    
    # === PASO 5: Clasificar el triángulo ===
    
    # --- Caso 1: Triángulo EQUILÁTERO ---
    # Todos los lados son exactamente iguales
    if lado1 == lado2 == lado3:
        print("\n" + "="*50)
        print("El triángulo es EQUILÁTERO")
        print(f"Los tres lados son iguales: {lado1}")
        print(f"El perímetro es: {perimetro}")
    
    # --- Caso 2: Triángulo ESCALENO ---
    # Todos los lados son diferentes
    elif lado1 != lado2 and lado1 != lado3 and lado2 != lado3:
        print("\n" + "="*50)
        print("El triángulo es ESCALENO")
        print("Todos los lados son diferentes")
        
        # Encontrar el lado más largo
        # Usamos max() que devuelve el valor máximo
        lado_mas_largo = max(lado1, lado2, lado3)
        print(f"El lado más largo mide: {lado_mas_largo}")
        print(f"El perímetro es: {perimetro}")
    
    # --- Caso 3: Triángulo ISÓSCELES ---
    # Exactamente dos lados son iguales
    else:
        print("\n" + "="*50)
        print("El triángulo es ISÓSCELES")
        print("Tiene dos lados iguales y uno diferente")
        
        # Identificar cuál lado es el diferente
        if lado1 == lado2:
            print(f"Lados iguales: {lado1} y {lado1}")
            print(f"Lado diferente: {lado3}")
        elif lado1 == lado3:
            print(f"Lados iguales: {lado1} y {lado1}")
            print(f"Lado diferente: {lado2}")
        else:  # lado2 == lado3
            print(f"Lados iguales: {lado2} y {lado2}")
            print(f"Lado diferente: {lado1}")
        
        print(f"El perímetro es: {perimetro}")
    
    print("\n" + "="*50)
    print("¡Hasta la próxima!")


# === INICIAR EL PROGRAMA ===
# Esta línea ejecuta la función cuando se corre el archivo
if __name__ == "__main__":
    clasificar_triangulo()
```
