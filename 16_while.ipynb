# BUCLE WHILE

```python
# BUCLE WHILE

i = 1

while i < 10:
    print(i) # Si lo dejas hasta acá, nunca termina de ejecutarse y debes hacer CTRL + C para que finalice en la terminal. Necesitas terminar el While para que finalice. 
    i += 1 # Al agregar el contador, si se detiene el bucle hasta que se cumple la condición. No se imprime el 10. 
```

```python
i = 1

while i <= 10:
    print(i)
    i += 1 # Se imprime el 10 porque usas <=.
```

```python
# Break finaliza el bucle While sin que cumpla la condición.

i = 1

while i <= 10:
    print(i)
    if i == 5:
        break # Al llegar a 5 se finaliza el bucle. 
    i += 1
```

```python
# Pasar el print antes de aumentar el contador. 

i = 1

while i <= 10:
    if i == 5:
        break # Al llegar a 4 se finaliza el bucle. 
    print(i)
    i += 1
```

```python
# El print va debajo del aumento de i.

i = 1

while i <= 10:
    i += 1
    print(i) # Empieza en 2 y termina en 11. Porque al ser menos a 10, primero aumento y después imprimió.

# Se corrige si ponemos a i empezando en 0 y es solo menor a 10.

print("\n")

i = 0

while i < 10:
    i += 1
    print(i)
```

```python
# Se suma de a 2. En este caso sobra el if porque nunca usa el 5.

i = 0

while i <= 10:
    if i == 5:
        break # No tiene en cuenta el 5 y no aplica el break. 
    print(i)
    i += 2 # Solo ejecuta los pares. 
```

```python
# continuo: Hace que se salte la instrucción.
# Se debe agregar el contador antes del if y el print o entra en un bucle infinito.  

i = 0

while i < 10:
    i += 1
    if i == 5:
        continue # Salta el 5. 
    print(i)
```

```python
# else: solo funciona cuando se acaba la condición del bucle.

i = 0

while i < 10:
    i += 1
    if i == 5:
        continue # Salta el 5. 
    print(i)
else:
    print("i dejó de ser menor a 10") # Se imprime después de llegar a 10. 
```
